# Combine them

```python
def try_parsing_date(text):
    for fmt in ('%Y-%m-%d %H:%M:%S', '%Y-%m-%d'):
        try:
            return datetime.strptime(text, fmt)
        except ValueError:
            pass
    raise ValueError('No valid date format found.')

def reverse_csv(path):
    df = pd.read_csv(path)
    first_str = df.iloc[0][0]
    first = try_parsing_date(first_str)
    last_str = df.iloc[len(df)-1][0]
    last = try_parsing_date(last_str)
    if first > last:
        df = df.iloc[::-1]
        df.to_csv(path, index=False)
    else:
        print(f'{path} is already in the correct order')
```

```python
dir = './data/daily/'
i = 0
n = len(os.listdir(dir))
for file in os.listdir(dir):
    if file == '.DS_Store':
        continue
    path = dir + file
    reverse_csv(path)
    i += 1
    print(f'{i}/{n} file {file} processed')
print("Reverse Successful")
```

```python
path = './data/dates/bull.csv'
def convert_date_format(s):
    ll = s.split('/')
    assert len(ll) == 3
    return f'{ll[2]}-{ll[0]}-{ll[1]}'
df = pd.read_csv(path)
df = df.applymap(convert_date_format)
df.to_csv(path, index=False)
df
```
